```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (Ensure loan_approval_data.csv is uploaded)
df = pd.read_csv('/content/drive/MyDrive/ML Coursework/loan_approval_data (1).csv')

# 1. Check the dimensions of the data
print(f"Dataset Shape: {df.shape}")

# 2. Check data types and non-null counts
print("\n--- Data Info ---")
print(df.info())

# 3. Check for missing values in each column
print("\n--- Missing Values ---")
print(df.isnull().sum())

# 4. Descriptive Statistics for numerical columns
print("\n--- Descriptive Statistics ---")
display(df.describe())
```

```python
# Set the visual style
sns.set(style="whitegrid")

# Create a count plot for loan_approval_status
plt.figure(figsize=(8, 5))
ax = sns.countplot(x='loan_approval_status', data=df, palette='viridis')

# Adding labels and title
plt.title('Distribution of Loan Approval Status', fontsize=15)
plt.xlabel('Approval Status (0 = Rejected, 1 = Approved)', fontsize=12)
plt.ylabel('Count of Clients', fontsize=12)

# Display percentages on top of bars (High mark tip for presentation)
total = len(df['loan_approval_status'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2 - 0.05
    y = p.get_height()
    ax.annotate(percentage, (x, y), size = 12, va='bottom')

plt.show()
```
